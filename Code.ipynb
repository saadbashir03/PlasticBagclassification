```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras import layers

# Set the path to your dataset
dataset_path = '/content/Data'

# Load the dataset
train_dataset = image_dataset_from_directory(
    dataset_path,
    image_size=(256, 256),  # Resize the images to a uniform size
    batch_size=32,
    label_mode='categorical'  # Use categorical labels for multi-class classification
)

# Split into training and validation sets
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset = train_dataset.take(train_size)
val_dataset = train_dataset.skip(train_size)

# Define the data augmentation layer
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),  # Randomly flip images horizontally and vertically
    layers.RandomRotation(0.2),  # Randomly rotate images by up to 20%
    layers.RandomZoom(0.2),  # Randomly zoom into images
    layers.RandomContrast(0.2)  # Randomly adjust contrast
])

# Apply data augmentation during training
augmented_train_dataset = train_dataset.map(
    lambda x, y: (data_augmentation(x, training=True), y)
)

# You can now train your model with augmented_train_dataset
```

```python
import os
import zipfile
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing import image_dataset_from_directory

# Set the path to your dataset
dataset_path = '/content/Data'

# Step 2: Create an ImageDataGenerator to load and preprocess the images
train_dir = os.path.join(dataset_path)

# ImageDataGenerator for data augmentation and normalization
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# Step 3: Build the CNN model
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(3, activation='softmax')  # Output layer for 3 categories (Garbage, Paper, Plastic)
])

# Step 4: Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Step 5: Train the model
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=validation_generator
)

# Step 6: Plot training results
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(10)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

The training and validation accuracy graph demonstrates a steady improvement in both training and validation accuracy over the epochs. By the final epoch, the training accuracy is around 85%, and the validation accuracy also approaches a similar value, which indicates that the model is learning well without significant overfitting.

However, looking at the training and validation loss, the validation loss starts higher and decreases but exhibits some fluctuations, especially in later epochs. This could indicate the model experiencing slight instability during validation, possibly due to overfitting or noise in the validation set.

### Key Observations:
- **Training Accuracy**: It consistently improves, indicating that the model is learning and fitting the training data well.
- **Validation Accuracy**: Shows a similar trend to training accuracy, indicating good generalization with slight fluctuations.
- **Training Loss**: Decreases steadily, which is expected as the model optimizes.
- **Validation Loss**: Fluctuates towards the end, which could be an indication that the model might be overfitting, or that the validation set is challenging.

In summary, the model performs well, but the fluctuations in validation loss suggest a potential need for further tuning (such as adjusting dropout rates, regularization, or learning rate) to stabilize and improve generalization.

```python
import numpy as np
import sklearn
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Make predictions on the validation set
validation_steps = validation_generator.samples // validation_generator.batch_size
predictions = model.predict(validation_generator, steps=validation_steps)

# Step 2: Get the predicted class labels and true labels
predicted_classes = np.argmax(predictions, axis=1)
true_classes = validation_generator.classes
class_labels = list(validation_generator.class_indices.keys())

# Step 3: Generate the confusion matrix
conf_matrix = confusion_matrix(true_classes[:len(predicted_classes)], predicted_classes)

# Step 4: Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
import numpy as np
import ipywidgets as widgets
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from PIL import Image
import io

# Load your saved model
model = load_model('/content/garbage_classification_model.h5')

# Define a mapping from indices to class names (replace with your class names)
class_names = ['Garbage', 'Paper', 'Plastic']

# Function to preprocess image and predict
def predict_image(img):
    img = img.resize((150, 150))  # Resize the image to match the model's input shape
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array /= 255.0  # Rescale the image to match preprocessing
    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions, axis=1)
    return class_names[predicted_class[0]]

# Function to handle file upload and prediction
def on_image_upload(change):
    # Get the uploaded image
    uploaded_file = change['new']
    img_data = uploaded_file[list(uploaded_file.keys())[0]]['content']

    # Convert to a PIL image and display it
    img = Image.open(io.BytesIO(img_data))
    display(img)

    # Predict and display the result
    prediction = predict_image(img)
    print(f"Predicted Class: {prediction}")

# Create a file upload widget
upload_widget = widgets.FileUpload(
    accept='image/*',  # Accept only images
    multiple=False  # Single file upload
)

# Attach the function to the widget
upload_widget.observe(on_image_upload, names='value')

# Display the widget
display(upload_widget)
```
